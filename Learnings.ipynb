```python
#Functions
def abc():
    #non parameterised
    print("hello")
def bca(name):
    #parameterised
    print("hii",name)
def cab(name="pavneet"):
    #default parameterised
    print("hii",name)
def cba():
    print("Hola Amigo")
    cab()
abc()
bca("pavi")
cab()
cba()

#nested Functions
def parent():
    print("Bacche Ka Baapu")
    def child():
        print("I'm just a Baby...")
    return child()

parent()
```

```python
#keyword argument
#aah(name="hola",age="10")

#arbitary arguments(*args) stores multiple values
#arbitary keyword arguments(**kargs)
def profile(**info):
    for key,value in info.items():
        print(f"{key}: {value}")

profile(name="pavneet",age=18)
```
