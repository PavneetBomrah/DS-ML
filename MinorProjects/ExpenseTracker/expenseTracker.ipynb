```python
import numpy as np
import csv
theData = []
def load_data():
    try:
        with open("eggs.csv", 'r', newline='') as csvfile:
            reader = csv.reader(csvfile)
            theData = list(reader)
            return theData
    except FileNotFoundError:
        print("No Data Found")
        return [[]]


while True:
    print("\n"*2,"-" * 30,"\n"*2)
    print("\tWelcome to The Expense Tracker\n\n")
    choice = int(input("Enter task to perform\n\t1. Show Data\n\t2. Add/Edit Data\n\t3. Show Basic Report\n\t0. Exit\n"))
    print("\n"*2,"-" * 30,"\n"*2)
    match choice:
        case 1:
            print("\n"*2,"-" * 30,"\n"*2)
            print("Loaded Data: \n")
            print("\tCategory - Amount\n")
            data = load_data()
            for items in data:
                if len(items) < 2:
                    continue
                print(f"\t{items[0]} - {items[1]}")
            print("\n"*2,"-" * 30,"\n"*2)
        case 2:
            print("\n"*2,"-" * 30,"\n"*2)
            print("Altering Data...:")
            print("\n\tCategory - Amount\n")
            data = load_data()
            for items in data:
                if len(items) < 2:
                    continue
                print(f"\t{items[0]} - {items[1]}")

            category = input("\n\n  Enter the Category name you \n  would like to edit or enter \n  name for new\n ")

            found = False
            for d in data:
                if len(d)>=2 and d[0] == category:
                    found = True
                    break

            if found:
                for d in data:
                    if d[0] == category:
                        newValue = int(input(f"Enter new value for {category}"))
                        d[1] = newValue
                        break
            else:
                print("Category doesnt Exists creating new...\n")
                newValue = int(input(f"Enter new value for {category}"))
                data.append([category, newValue])
                        
            try:
                with open("eggs.csv", 'w', newline='') as csvfile:
                    writer = csv.writer(csvfile)
                    writer.writerows(data)
                success = True
            except Exception:
                print("An error occurred while writing.")
                success = False
            finally:
                if success:
                    print("Write Successful")
                else:
                    print("Write Failed")

            print("\n"*2,"-" * 30,"\n"*2)

        case 3:
            print("\n"*2,"-" * 30,"\n"*2)
            print("Basic Report: \n")
            data = load_data()
            categories = []
            amount = []
            for d in data:
                if len(d)>=2:
                    categories.append(d[0])
                    amount.append(int(d[1]))
            total = np.sum(amount)
            average = total/(len(amount)+1)
            maximum = [categories[amount.index(np.max(amount))],int(np.max(amount))]
            minimum = [categories[amount.index(np.min(amount))],int(np.min(amount))]
            
            print(f"Total = {total}")
            print(f"Average = {average}")
            print(f"Maximum = {maximum[0],maximum[1]}")
            print(f"Minimum = {minimum[0],minimum[1]}")
            print("\n"*2,"-" * 30,"\n"*2)

        case _:
            break
```
