```python
import numpy as np

# Monthly categories
categories = ['Rent', 'Groceries', 'Utilities', 'Transport', 'Entertainment', 'Health', 'Other']

# Monthly expenses for each category (in INR for example)
# Let's assume these are for one month
expenses = np.array([15000, 4200, 2200, 1200, 3000, 1100, 800])

def generate_report(categories, expenses):
    total_expense = np.sum(expenses)
    average_expense = np.mean(expenses)
    max_index = np.argmax(expenses)
    min_index = np.argmin(expenses)

    print("📊 Monthly Expense Report")
    print("-" * 30)
    
    # Print each category with corresponding expense
    for i in range(len(categories)):
        print(f"{categories[i]:<15}: ₹{expenses[i]}")
    
    print("-" * 30)
    print(f"Total Expense       : ₹{total_expense}")
    print(f"Average Expense     : ₹{average_expense:.2f}")
    print(f"Highest Spending    : {categories[max_index]} (₹{expenses[max_index]})")
    print(f"Lowest Spending     : {categories[min_index]} (₹{expenses[min_index]})")
    print("-" * 30)

# Optional: Loop to update expenses if you want user input (commented out for now)
# for i in range(len(categories)):
#     expenses[i] = float(input(f"Enter amount spent on {categories[i]}: "))

# Generate the report
generate_report(categories, expenses)
```
